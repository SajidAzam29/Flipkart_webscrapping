```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import csv
import requests
```

```python
url="https://www.flipkart.com/search?q=samsung+mobiles&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_8_na_na_ps&otracker1=AS_Query_OrganicAutoSuggest_3_8_na_na_ps&as-pos=3&as-type=RECENT&suggestionId=samsung+mobiles&requestId=2ab8776c-4c04-4469-8636-3919f3c5649c&as-searchtext=Samsung+&p%5B%5D=facets.price_range.from%3D10000&p%5B%5D=facets.price_range.to%3D20000"
```

```python
req=requests.get(url)
```

```python
content=BeautifulSoup(req.content,'html.parser')
```

```python
print(content)
```

```python
name=content.find_all('div',{'class':'_4rR01T'})
price=content.find_all('div',{'class':'_30jeq3 _1_WHN1'})
rating=content.find_all('div',{'class':'_3LWZlK'})
```

```python
print(name[0].text)
print(price[0].text)
print(rating[0].text)
```

```python
n=[]
p=[]
r=[]
```

```python
for i in name:
    n.append(i.text)
for i in price:
    p.append(i.text)
for i in  rating:
    r.append(i.text)
```

```python
data={'name':n , 'price':p , 'rating':r}
df=pd.DataFrame(data)
```

```python
print(df)
```

```python
df.to_csv("Flipkart web scrapping.csv")
```
